```python
#count gender
ax=sns.countplot(x='Gender',data=df,color='blue')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Boxplot to visualize Exam_Score distribution by Gender
plt.figure(figsize=(6, 4))
sns.boxplot(x='Gender', y='Exam_Score', data=df)
plt.title('Boxplot of Exam Scores by Gender')
plt.show()
```

```python
 # Tutoring Sessions vs Exam Performance
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Tutoring_Sessions', y='Exam_Score', hue='Gender', style='Attendance')
plt.title("Tutoring_Sessions vs Exam Score")
plt.xlabel("Tutoring_Sessions")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

```python
#Bar plot to visualize the average Exam_Score by School Type
plt.figure(figsize=(6, 4))
sns.barplot(x='School_Type', y='Exam_Score', data=df,color='green')
plt.title('Average Exam Scores by School Type')
plt.show()
```

```python
#Scatter plot to visualize relationship between Hours_Studied and Exam_Score
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Scatter Plot of Hours Studied vs Exam Score')
plt.show()
```

```python
#Bar plot to visualize the School Type by family income
plt.figure(figsize=(6, 4))
sns.barplot(x='School_Type', y='Family_Income', data=df)
plt.title('School Type by family income')
plt.show()
```

```python
# Exam performance based on attendance
import matplotlib.pyplot as plt
attendance_labels = ['High Attendance', 'Medium Attendance', 'Low Attendance']
attendance_sizes = [50, 30, 20]
colors = ['#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0)
plt.figure(figsize=(8, 6))
plt.pie(attendance_sizes, labels=attendance_labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
plt.title("Exam Performance Distribution by Attendance")
plt.axis('equal')
plt.show()

```

```python
#Histogram to visualize the distribution of Exam Scores
plt.figure(figsize=(6, 4))
sns.histplot(df['Exam_Score'], bins=10, kde=True)
plt.title('Histogram of Exam Scores')
plt.show()
```

```python
# Parental involvement analysis
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Parental_Involvement', y='Exam_Score', ci=None)
plt.title("Parental Involvement and Exam Performance")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
#Line plot to visualize trend of Exam Scores over Previous Scores
plt.figure(figsize=(6, 4))
sns.lineplot(x='Previous_Scores', y='Exam_Score', data=df, marker='o')
plt.title('Line Plot of Exam Scores vs Previous Scores')
plt.show()
```

```python
# Analyzing the effect of studies
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Hours_Studied', bins=10, hue='Exam_Score', kde=True)
plt.title("Distribution of Study Hours by Exam score")
plt.xlabel("Hours Studied")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot to visualize Exam Scores by Access to Resources
plt.figure(figsize=(6, 4))
sns.boxplot(x='Access_to_Resources', y='Exam_Score', data=df)
plt.title('Boxplot of Exam Scores by Access to Resources')
plt.show()
```

```python
#Top 10 Students by Exam Score:
top_students = df.nlargest(10, 'Exam_Score')
print("Top 10 Students by Exam Score:")
print(top_students[[ 'Exam_Score', 'Hours_Studied', 'Attendance']])
```

```python
 #Attendance Impact on Pass/Fail Rates
df['Pass'] = df['Exam_Score'] >= 50  # Assuming 50 is the passing score

attendance_pass_rate = df.groupby('Attendance')['Pass'].mean() * 100
plt.figure(figsize=(8, 6))
attendance_pass_rate.plot(kind='bar', color='cornflowerblue')
plt.title("Pass Rate by Attendance Level")
plt.xlabel("Attendance Level")
plt.ylabel("Pass Rate (%)")
plt.show()
```
