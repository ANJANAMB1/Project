```python
 import numpy as np
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
```

```python
import pandas as pd
file_path ='/content/StudentPerformanceFactors.csv'
df = pd.read_csv(file_path)
df.shape
```

Data Cleaning

```python
#drop unrelated/blank columns
df.drop(['Grade','Marks' ],axis=1, inplace=True)
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
# Handling missing values
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
#count gender
ax=sns.countplot(x='Gender',data=df,color='blue')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Calculate total attendance by Gender
total_attendance = df.groupby('Gender')['Attendance'].sum().reset_index()

# Barplot for total attendance
plt.figure(figsize=(6, 4))
sns.barplot(x='Gender', y='Attendance', data=total_attendance, palette='muted')
plt.title('Total Attendance by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Attendance')
plt.ylim(0, total_attendance['Attendance'].max() + 10)
plt.show()
```

```python
 # Tutoring Sessions vs Exam Performance
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Tutoring_Sessions', y='Exam_Score', hue='Gender', style='Attendance')
plt.title("Tutoring_Sessions vs Exam Score")
plt.xlabel("Tutoring_Sessions")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

```python
#Line plot to visualize the average Exam_Score by School Type

plt.figure(figsize=(6, 4))
sns.lineplot(x='School_Type', y='Exam_Score', data=df, color='green', marker='o')
plt.title('Exam Scores by School Type')
plt.xlabel('School Type')
plt.ylabel('Exam Score')
plt.show()
```

```python
#Scatter plot to visualize relationship between Hours_Studied and Exam_Score
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.title('Scatter Plot of Hours Studied vs Exam Score')
plt.show()
```

```python
# Calculate total attendance by Distance_From_Home
total_attendance = df.groupby('Distance_From_Home')['Attendance'].sum().reset_index()

# Barplot for total attendance by distance
plt.figure(figsize=(6, 4))
sns.barplot(x='Distance_From_Home', y='Attendance', data=total_attendance, palette='viridis')
plt.title('Total Attendance by Distance from Home')
plt.xlabel('Distance from Home')
plt.ylabel('Total Attendance')
plt.show()
```

```python
#Exam performance based on attendance
attendance_labels = ['High Attendance', 'Medium Attendance', 'Low Attendance']
attendance_sizes = [50, 30, 20]
colors = ['#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0)
plt.figure(figsize=(8, 6))
plt.pie( attendance_sizes,labels=attendance_labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
plt.title("Exam Performance Distribution by Attendance")
plt.axis('equal')
plt.show()
```

```python
#Scatterplot to visualize the Exam Scores related to learning disabilities
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Exam_Score', y='Learning_Disabilities')
plt.title('Exam Scores related to learning disabilities')
plt.show()
```

```python
#Parental involvement analysis
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Parental_Involvement', y='Internet_Access')
plt.title("Parental Involvement and Internet_Access")
plt.xlabel("Parental Involvement")
plt.ylabel("Internet_Access")
plt.show()
```

```python
#Line plot to visualize trend of Exam Scores over Previous Scores
plt.figure(figsize=(6, 4))
sns.lineplot(x='Previous_Scores', y='Exam_Score', data=df, marker='o')
plt.title('Line Plot of Exam Scores vs Previous Scores')
plt.show()
```

```python
#Percentage of Students with and without Internet Access
internet_access_counts = df['Internet_Access'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(internet_access_counts, labels=internet_access_counts.index, autopct='%1.1f%%',startangle=90,colors=['skyblue', 'orange'])
plt.title('Percentage of Students with and without Internet Access')
plt.show()
```

```python
#Boxplot to visualize Exam Scores by Access to Resources
plt.figure(figsize=(6, 4))
sns.boxplot(x='Access_to_Resources', y='Exam_Score', data=df)
plt.title('Boxplot of Exam Scores by Access to Resources')
plt.show()
```

```python
#Top 10 Students by Exam Score:
top_students = df.nlargest(10, 'Exam_Score')
print("Top 10 Students by Exam Score:")
print(top_students[[ 'Exam_Score', 'Hours_Studied', 'Attendance']])
```

```python
#Attendance Impact on Pass/Fail Rates
import matplotlib.pyplot as plt
df['Pass'] = df['Exam_Score'] >= 50

attendance_pass_rate = df.groupby('Attendance')['Exam_Score'].mean() * 100
plt.figure(figsize=(8, 4))
plt.plot(attendance_pass_rate.index, attendance_pass_rate.values, marker='o', color='cornflowerblue', linestyle='-', linewidth=2)
plt.title("Attendance Impact on Pass/Fail Rates")
plt.xlabel("Attendance Level")
plt.ylabel("Pass Rate (%)")
plt.grid(True)
plt.show()
```
