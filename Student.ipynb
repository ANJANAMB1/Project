```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
 # Hours Studied vs Exam Performance
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Hours_Studied', y='Exam_Score', hue='Gender', style='Attendance')
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

```python
# Exam performance based on attendance
import matplotlib.pyplot as plt

attendance_labels = ['High Attendance', 'Medium Attendance', 'Low Attendance']
attendance_sizes = [50, 30, 20]
colors = ['#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0)

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(attendance_sizes, labels=attendance_labels, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode)
plt.title("Exam Performance Distribution by Attendance")
plt.axis('equal')
plt.show()

```

```python
# Parental involvement analysis
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Parental_Involvement', y='Exam_Score', ci=None)
plt.title("Parental Involvement and Exam Performance")
plt.xlabel("Parental Involvement")
plt.ylabel("Average Exam Score")
plt.show()
```

```python
# Analyzing the effect of studies
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Hours_Studied', bins=10, hue='Exam_Score', kde=True)
plt.title("Distribution of Study Hours by Exam score")
plt.xlabel("Hours Studied")
plt.ylabel("Frequency")
plt.show()
```

```python
# Gender-wise performance comparison
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='Gender', y='Exam_Score', split=True, inner='quartile')
plt.title("Gender-wise Exam Score Distribution")
plt.xlabel("Gender")
plt.ylabel("Exam Score")
plt.show()
```

```python
#Relationship Between Hours Studied and Exam Score
plt.figure(figsize=(8, 6))
sns.regplot(data=df, x='Hours_Studied', y='Exam_Score', scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title("Relationship Between Hours Studied and Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```

```python
top_students = df.nlargest(10, 'Exam_Score')
print("Top 10 Students by Exam Score:")
print(top_students[[ 'Exam_Score', 'Hours_Studied', 'Attendance']])
```

```python
 #Attendance Impact on Pass/Fail Rates
df['Pass'] = df['Exam_Score'] >= 50  # Assuming 50 is the passing score

attendance_pass_rate = df.groupby('Attendance')['Pass'].mean() * 100
plt.figure(figsize=(8, 6))
attendance_pass_rate.plot(kind='bar', color='cornflowerblue')
plt.title("Pass Rate by Attendance Level")
plt.xlabel("Attendance Level")
plt.ylabel("Pass Rate (%)")
plt.show()
```
